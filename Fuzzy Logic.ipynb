```python
def triangular(x, a, b, c):
    return np.maximum(np.minimum((x - a) / (b - a), (c - x) / (c - b)), 0)
```

```python
def fuzzify_dirt_level(dirt_level):
    light = triangular(dirt_level, 0, 0.3, 0.6)
    moderate = triangular(dirt_level, 0.3, 0.6, 0.9)
    heavy = triangular(dirt_level, 0.6, 0.9, 1)
    return {'Light': light, 'Moderate': moderate, 'Heavy': heavy}
```

```python
def fuzzify_load_size(load_size):
    small = triangular(load_size, 0, 0.4, 0.7)
    medium = triangular(load_size, 0.4, 0.7, 1)
    large = triangular(load_size, 0.7, 1, 1.2)
    return {'Small': small, 'Medium': medium, 'Large': large}
```

```python
def fuzzify_fabric_type(fabric_type):
    delicate = triangular(fabric_type, 0, 0.3, 0.7)
    normal = triangular(fabric_type, 0.3, 0.7, 1)
    heavy = triangular(fabric_type, 0.7, 1, 1.2)
    return {'Delicate': delicate, 'Normal': normal, 'Heavy': heavy}
```

```python
def apply_rules(dirt_level_fuzzy, load_size_fuzzy, fabric_type_fuzzy):
    rules = []
    very_short_time = min(dirt_level_fuzzy['Light'], load_size_fuzzy['Small'], fabric_type_fuzzy['Delicate'])
    rules.append(('Very Short', very_short_time))

    very_long_time = min(dirt_level_fuzzy['Heavy'], load_size_fuzzy['Large'], fabric_type_fuzzy['Heavy'])
    rules.append(('Very Long', very_long_time))

    medium_time = min(dirt_level_fuzzy['Moderate'], load_size_fuzzy['Medium'], fabric_type_fuzzy['Normal'])
    rules.append(('Medium', medium_time))

    return rules
```

```python
def defuzzify(rules):
    time_ranges = {
        'Very Short': 15,
        'Short': 30,
        'Medium': 45,
        'Long': 60,
        'Very Long': 90
    }

    numerator = 0
    denominator = 0

    for (time_label, membership_value) in rules:
        if membership_value > 0:
            numerator += membership_value * time_ranges[time_label]
            denominator += membership_value

    if denominator == 0:
        return 0
    return numerator / denominator
```

```python
def fuzzy_washing_machine_controller(dirt_level, load_size, fabric_type):
    dirt_level_fuzzy = fuzzify_dirt_level(dirt_level)
    load_size_fuzzy = fuzzify_load_size(load_size)
    fabric_type_fuzzy = fuzzify_fabric_type(fabric_type)

    rules = apply_rules(dirt_level_fuzzy, load_size_fuzzy, fabric_type_fuzzy)
    washing_time = defuzzify(rules)
    return washing_time
```

```python
def plot_membership_functions(dirt_level, load_size, fabric_type):
    x_dirt = np.linspace(0, 1, 100)
    x_load = np.linspace(0, 1, 100)
    x_fabric = np.linspace(0, 1, 100)

    # Dirt level membership
    light = triangular(x_dirt, 0, 0.3, 0.6)
    moderate = triangular(x_dirt, 0.3, 0.6, 0.9)
    heavy = triangular(x_dirt, 0.6, 0.9, 1)

    plt.figure(figsize=(15, 5))

    plt.subplot(1, 3, 1)
    plt.plot(x_dirt, light, label='Light', color='blue')
    plt.plot(x_dirt, moderate, label='Moderate', color='orange')
    plt.plot(x_dirt, heavy, label='Heavy', color='red')
    plt.axvline(dirt_level, color='green', linestyle='--', label='Input Dirt Level')
    plt.title('Dirt Level Membership Functions')
    plt.xlabel('Dirt Level')
    plt.ylabel('Membership Degree')
    plt.legend()
    plt.grid()

    small = triangular(x_load, 0, 0.4, 0.7)
    medium = triangular(x_load, 0.4, 0.7, 1)
    large = triangular(x_load, 0.7, 1, 1.2)

    plt.subplot(1, 3, 2)
    plt.plot(x_load, small, label='Small', color='blue')
    plt.plot(x_load, medium, label='Medium', color='orange')
    plt.plot(x_load, large, label='Large', color='red')
    plt.axvline(load_size, color='green', linestyle='--', label='Input Load Size')
    plt.title('Load Size Membership Functions')
    plt.xlabel('Load Size')
    plt.ylabel('Membership Degree')
    plt.legend()
    plt.grid()

    delicate = triangular(x_fabric, 0, 0.3, 0.7)
    normal = triangular(x_fabric, 0.3, 0.7, 1)
    heavy = triangular(x_fabric, 0.7, 1, 1.2)

    plt.subplot(1, 3, 3)
    plt.plot(x_fabric, delicate, label='Delicate', color='blue')
    plt.plot(x_fabric, normal, label='Normal', color='orange')
    plt.plot(x_fabric, heavy, label='Heavy', color='red')
    plt.axvline(fabric_type, color='green', linestyle='--', label='Input Fabric Type')
    plt.title('Fabric Type Membership Functions')
    plt.xlabel('Fabric Type')
    plt.ylabel('Membership Degree')
    plt.legend()
    plt.grid()

    plt.tight_layout()
    plt.show()
```
